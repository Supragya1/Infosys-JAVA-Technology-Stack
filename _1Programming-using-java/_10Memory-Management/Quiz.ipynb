Which of the following statements are true?

- You cannot be certain at what point of time garbage collection will occur [x]
- An object without reference will be immediately garbage collected []
- An object will be eligible for garbage collection after the execution of the method is over [x]
- Garbage collection ensures programs will never run out of memory []

Explanation :
Once an object is not needed, it will be subject to garbage collection but you cannot be certain at what point of time will garbage collection happen.

Explanation :
A reference variable will be removed from the stack as soon as the method finishes execution. If this reference was referring to an object, that object becomes eligible for garbage collection.

- 0 []
