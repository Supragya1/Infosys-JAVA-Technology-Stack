**Q1 of 4**

What will be the output of the below code?

``` java
class Demo {
	public int value = 20;

	Demo() {
		value = 40;
	}
}

public class Tester {
	public static void main(String args[]) {
		Demo demo = new Demo();
		System.out.println(demo.value);
	}
}
```
- 0 []
- 20 []
- 40 [x]
- Compilation error: variable 'value' is not declared []
  
Explanation :

When the object of Demo class is created, constructor gets invoked. So, the value of the instance variable 'value' is 40.

**Q2 of 4**

What will be the output of the below code?

``` java
class Demo {
	public int var1 = 20;
	public int var2 = 40;

	Demo(int v1, int v2) {
		var1 = v1;
		var2 = v2;
	}
}

public class Tester {
	public static void main(String args[]) {
		Demo demo = new Demo();
		System.out.println(demo.var1);
		System.out.println(demo.var2);
	}
}
```

- 20 40 []
- 0 0 []
- 40 20 []
- compilation error: class Demo has no default constructor [x]
  
Explanation :

If a parameterized constructor is created by the developer, Java does not supply a default constructor. Since, there is no parameterless constructor present, there will be a compilation error.
