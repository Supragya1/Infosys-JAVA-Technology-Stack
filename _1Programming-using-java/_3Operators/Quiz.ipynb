What will be the output of the below code? 

``` java
public class Demo {
	public static void main(String args[]) {
		int a = 10;
		int b = 2;
		System.out.println((a < b) ? a++ : --b);
	}
}
```
- 2 []
- 1 [x]
- 10 []
- 11 []
  
Explanation :

postfix operator first uses the value and then increments it and prefix operator first increments the value and then uses it

**Q2 of 3**

What will be the output of the below code?

``` java
public static void main(String args[]) { 
	int i = 10; 
	if(i) {
		System.out.println("Hello"); 
	} 
	else { 
		System.out.println("Bye"); 
	}
}
```
