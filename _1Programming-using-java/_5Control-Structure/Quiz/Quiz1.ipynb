**Q8 of 14**

What will be the output of the below program?

```java
public class Demo {
	public static void main(String[] args) {
		int i = -1, j = -1;
		switch (i) {
		case -1:
			j = 1;
		case 2:
			j = 2;
			break;
		default:
			j = 0;
		}
		System.out.println("j = " + j);
	}
}
```

- j = 2 [x] 
- j = 1 []
- j = 0 []
- Compilation Error due to absence of break in first case block []

Explanation :

Since i=-1, the first case gets matched and gets executed but since there is no break statement present in the first case, the second case also gets executed and the value of j gets updated to 2. Then, the control comes out of switch block after encountering the break statement.

once the case condition is true, all the statements below it get executed until it encounters the break statements or reaches the end of the switch block. Therefore the second case also gets executed and j gets updated as j= 2. Hence this is the answer.
