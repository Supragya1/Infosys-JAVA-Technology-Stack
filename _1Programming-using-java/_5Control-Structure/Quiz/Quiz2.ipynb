What is the output of the following code?

```java
public class Demo {
	public static void main(String[] args) {
		int i = 1, j = 10;
		do {
			j--;
		} while (++i < 5);
		System.out.println("i = " + i + " and j = " + j);
	}
}
```
- i = 5 and j = 6
- i = 5 and j = 5
- i = 6 and j = 6
- i = 6 and j = 5
  
Explanation :

j gets decremented by 1 and i gets incremented by 1 in every iteration of the loop.
After the execution of the loop, the value of i and j will be 5 and 6 respectively.

**Q7 of 10**

What is the output of the following code?

```java
public class Demo {
	public static void main(String[] args) {
		int counter = 0;
		while (counter <= 7) {
			if (counter % 2 == 0) {
				counter += 1;
				continue;
			} else {
				counter += 1;
			}
			System.out.println(counter);
		}
	}
}
```

- 1 3 5 7 []
- 2 4 6 []
- 1 3 5 []
- 2 4 6 8 [x] 
  
Explanation :

For each iteration of the while loop, for each of the even value of counter then the condition counter%2==0 evaluates to true and counter gets incremented by one skipping rest of the statements of the loop because of continue statement. 
Whenever the value of counter is an odd number, the else block gets executed where the counter gets incremented by 1. Then, the value of counter gets displayed. Finally, 2, 4, 6 and 8 get displayed.

- 5 and 2 []
- 8 and 5 []
- 6 and 3 [x]
- 6 and 4 []
  
Explanation :

Whenever there is && operator, if the first expression evaluates to false, the second expression does not even get evaluated.
So, the second expression gets evaluated only when the first expression evaluates to true, i.e., when the value of num1 is 3.
