**Q5 of 7**

What will be the output of the following code snippet?

``` java
class A {
	public int method1() {
		return 20;
	}
}

class B extends A {
	public int method1() {
		return 30;
	}

	public int method2() {
		return 40;
	}
}

class C extends B {
	public int method2() {
		return 20;
	}
}

public class Tester {
	public static void main(String[] args) {
		A a = new A();
		B b = new B();
		C c = new C();
		System.out.println(a.method1() + c.method1() + c.method2());
	}
}   
```

- 90 []
- 80 []
- 70 [x]
- Error : method1() should be overridden in class C []

Explanation :
For a.method1(), method1 of class A is invoked.
For c.method1(), method1 of class B is invoked.
For c.method2(), method2 of class C is invoked. 
So, the output is 20+30+20=70.
Line 3, 4 and 5 create objects of class A (a), class B (b) and class C (c) respectively. Line 6 invokes.
a.method1() which takes control to method1() of class A is returning 20
c.method1() which takes control to method1() of class B is returning 30 &
c.method2() which takes control to method2() of class C is returning 20.
Hence this is the answer.
Note: When an overridden method is invoked, which method will be invoked depends on the object type. If the object belongs to child class, child's method will be invoked. If it belongs to parent class, parent's method will be invoked.

**Q6 of 7**

What will be the output of the following code snippet?

``` java
class Loan {
	int interestRate;

	Loan(int interestRate) {
		this.interestRate = interestRate;
		System.out.println("Request for loan");
	}
}

public class HomeLoan extends Loan {
	int interestRate;

	HomeLoan() {
		super(10);
		interestRate = super.interestRate;
		System.out.println("Request for home loan " + interestRate);
	}

	public static void main(String args[]) {
		HomeLoan homeLoan = new HomeLoan();
	}
}
```
