**Q2 of 3**

What will be the output of the code given below?

```java
public class Employee {
	private String name;
	private int empId;
	private String city;

	Employee() {
		this.city = "New York";
	}

	Employee(String name, int empId) {
		this.name = name;
		this.empId = empId;
	}

	public static void main(String[] args) {
		Employee employee1 = new Employee("John", 101);
		Employee employee2 = new Employee();

		System.out.println(employee1.name + " " + employee1.empId + " "
				+ employee1.city);
		System.out.println(employee2.name + " " + employee2.empId + " "
				+ employee2.city);
	}
}
```

- John 101 New York John 101 New York []
- John 101 null null 0 New York [x]
- John 101 New York null 0 New York []
- John 101 null John 101 New York []
  
Explanation :
The parameterized constructor is invoked while creating employee1 and the parameterless constructor is invoked while creating employee2. In the parameterized constructor, city is not initialized and in the parameterless constructor, only city is initialized.
For employee1 object, the parameters are passed, hence parameterized constructor is invoked. Instance variable city is not initialized, hence the value for city is null.
For employee2 object, the default constructor is invoked. Instance variables name and empId are not initialized, hence the value for name and empId is null and 0 respectively. 

**Q3 of 3**
What will be the output of the code given below?

```java
public class Employee {
	private String name;
	private int empId;
	private String city;

	Employee() {
		this.city = "New York";
	}

	Employee(String name, int empId) {
		this();
		this.name = name;
		this.empId = empId;
	}

	public static void main(String[] args) {
		Employee employee1 = new Employee("John", 101);
		Employee employee2 = new Employee();

		System.out.println(employee1.name + " " + employee1.empId + " "
				+ employee1.city);
		System.out.println(employee2.name + " " + employee2.empId + " "
				+ employee2.city);
	}
}
```

- John 101 New York John 101 New York []
- John 101 null null 0 New York []
- John 101 New York null 0 New York  [x]
- John 101 null John 101 New York []
  
Explanation  :
The parameterized constructor is invoked while creating employee1. Also, the parameterless constructor is invoked from the parameterized constructor. So, for employee1, all the instance variables get initialized.
The parameterless constructor is invoked while creating employee2. So, for employee2, only city is initialized.
For employee1 object, the parameters are passed, hence parameterized constructor is invoked. Since "this()" is used, city also gets initialized.
For employee2 object, the default constructor is invoked. Instance variables name and empId are not initialized, hence the value for name and empId is null and 0 respectively.
