### Q1) How many instance variables and objects are there in the following code?

```java
public class Tester {

    public static void main(String[] args) {

        Book myBook = new Book();
        Book yourBook = new Book();
        Book newBook;
    }

}

class Book {

    public String name;
    public double price;
    public String author;
    public int stock;
}
````
- instance variables: 4, objects: 3 []
- instance variables: 2, objects: 4 []
- instance variables: 4, objects: 2 [x]
- instance variables: 3, objects: 4 []


**Explanation:**

Instance variables are declared inside a class and are used for storing values of objects. There are 4 instance variables namely name, price, author, and stock.

An object is an instance of a class created using the new keyword. There are 2 objects that are created inside the main method.

---

### Q2) What is the output of the following code?
```java
public class Test {
    public static void main(String[] args) {
        Demo test = new Demo();
        System.out.println(test.flag);
    }
}

class Demo {
    public boolean flag;
}
```
- 0 []
- true []
- false [x]
- No output will be displayed []

**Explanation:**

Instance variable of a class will be initialized to default value unless done by the programmer.

false is the default value for the boolean data type.
