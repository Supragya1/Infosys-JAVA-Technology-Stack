**Q3 of 3**

Predict the output of the following code.
```java
abstract class A {
	public void disp() {
		System.out.println("Concrete method of parent class");
	}

	abstract public void disp2();
}

class B extends A {
	public void disp2() {
		System.out.println("Overriding abstract method");
	}

	public static void main(String args[]) {
		B obj = new B();
		obj.disp();
	}
}
```

- Overriding abstract method []
- Concrete method of parent class [x] 
- Compilation error: abstract class cannot have non abstract methods []
- Runtime error: non abstract method of parent cannot be invoked using child object []

Explanation :
disp() method is invoked and it is inherited by the child class.
