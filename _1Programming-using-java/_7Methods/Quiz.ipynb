**Q3 of 5**

Identify the number of instances and local variables present in the Student class given below.
``` java
public class Student {
	public int rollNo;
	public String studentName;

	public void findGrade(int marks) {
		char grade;
		if (marks > 90)
			grade = 'A';
		else if (marks < 90 && marks > 80)
			grade = 'B';
		else if (marks < 80 && marks > 70)
			grade = 'C';
		else
			grade = 'D';
		System.out.println(grade);
	}
}
```
- 2 instances and 2 local variable [x]
- 3 instance variables, 1 local variables []
- 2 instance variables, 1 local variables []
- 4 instance variables, 0 local variables []

Explanation :

Instance variables are declared inside a class and are used for storing values of objects.
Local variables are variables declared inside a method. Even parameters are local variables.
rollNo and studentName are instance variables and marks and grade are local variables.


**Q4 of 5**

What will be the output of the below code?
``` java
public class Tester {

	public int subtract(int a, int b) {
		return (a - b);
	}

	public static void main(String args[]) {
		int a = 10;
		int b = 20;
		Tester tester = new Tester();
		int res = tester.subtract(b, a);
		System.out.println(res);
	}

}
```
