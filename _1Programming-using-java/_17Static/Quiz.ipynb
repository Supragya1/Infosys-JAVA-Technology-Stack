**Q1 of 6**

How many static variables, local variables and instance variables are present in the code given below?

```java
public class Car {
	private String vehicleNumber;
	private String color;
	private static int numberOfCars = 0;

	public Car(String vehicleNumber, String color) {
		this.vehicleNumber = vehicleNumber;
		this.color = color;
	}

	public void display(String name) {
		System.out.println("Owner's name: " + name);
		System.out.println("Vehicle number: " + vehicleNumber);
		System.out.println("Color: " + color);
	}
}
```
- static variables: 1; instance variables: 2; local variables: 1 []
- static variables: 3; instance variables: 0; local variables: 2 []
- static variables: 1; instance variables: 2; local variables: 3 [x] 
- static variables: 3; instance variables: 2; local variables: 3 []

Explanation :

Instance variables are the ones that are declared inside a class. Static variables are the ones that are declared inside a class with static keyword. Local variables are the ones which are declared inside a method or a constructor, even parameters are local variables.
Here, numberOfCars is a static variable, vehicleNumber and color are instance variables and there are 2 parameters in the constructor and 1 parameter in the display() method, so there are 3 local variables.

**Q5 of 6**

Predict the output of the code given below.

```java
class Printer {
	static {
		System.out.println("Static block in Printer class");
	}

	public static void display(String message) {
		System.out.println(message);
	}
}

public class Tester {
	public static int sampleVariable = 1;

	static {
		System.out.println("Static block in Tester class");
	}

	public static void main(String[] args) {
		sampleVariable++;
		Printer p = null;
		System.out.println("In main");
		p.display("The value of sample variable is: " + sampleVariable);
	}
}
```

- Static block in Printer class; Static block in Tester class; In main; The value of sample variable is: 2 []
- Static block in Tester class; Static block in Printer class; In main; The value of sample variable is: 2 []
- Static block in Tester class; In main; Static block in Printer class; The value of sample variable is: 2 [x]
- Static block in Tester class; In main; The value of sample variable is: 2; Static block in Printer class []
  
Explanation :

First, the Tester class is loaded, hence, the static block of Tester class gets executed first. Then, 'In main' is displayed. After that, the Printer class gets loaded when the display() method is invoked and so, the static block of Printer class gets executed. Finally, the message gets displayed using the display() method.
